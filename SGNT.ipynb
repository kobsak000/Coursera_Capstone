# 1. Geting data.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

GET TABLE FROM WIKI

```python
url = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(url.content, "html.parser")
data = soup.find("table",{"class":"wikitable sortable"})
table_rows = data.find_all('tr')
```

CONVERT TABLE TO DATAFRAME

```python
#Table to df
l=[]
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    l.append(row)
```

CLEANING DATA

```python
df = pd.DataFrame(l, columns=["Postcode", "Borough","Neighborhood"])
df = df[df.Borough !='Not assigned']
df = df.replace(to_replace='None', value=np.nan).dropna()
df['Neighborhood'] = df['Neighborhood'].str.replace("\n","")
df['Neighborhood'] = np.where(df['Neighborhood'] == 'Not assigned', df['Borough'],df['Neighborhood'])
df=df.groupby(['Postcode','Borough'])['Neighborhood'].apply(lambda x: ','.join(x)).reset_index()
df
```

# 3. Clustering

```python
#!conda install -c conda-forge folium=0.5.0 --yes
import folium

print('Folium installed and imported!')
```

```python
latitude = 43.806
longitude = -79.194
```

```python
# create map and display it
neighborhood_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# display the map of San Francisco
neighborhood_map
```

```python
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(data.Latitude, data.Longitude):
    incidents.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add incidents to map
neighborhood_map.add_child(incidents)
```
